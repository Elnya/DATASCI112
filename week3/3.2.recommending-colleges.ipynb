# Recommending Colleges

Sally visits Rice University in Houston, TX and falls in love with the school. What other schools should she apply to? In this case study, we will use the [College Scorecard data](https://collegescorecard.ed.gov/) to make recommendations

## Question 0

Read in the data from  https://datasci112.stanford.edu/data/college_attributes.csv, and find Rice in this data set. What row is it?

```python
import pandas as pd
df = pd.read_csv('data/college_attributes.csv')
df
```

```python
rice_df = df[df['Institution'].str.contains('Rice University', na=False)]
rice_df
```

## Question 1

Based on only the admission rate and the number of undergraduates, what are the 5 schools that are most similar to Rice?

```python
features = ['AdmissionRate', 'Undergraduates']
df[features]
```

```python
# get the standard deviation of each column
df_standardized = ((df[features] - df[features].mean(axis='rows')) / df[features].std(axis='rows'))
df_standardized
```

```python
# finding the most simillar institutions to Rice University
import numpy as np

diffs = df_standardized - df_standardized[features].loc[1432]
dists = np.sqrt((diffs ** 2).sum(axis='columns')).sort_values()
dists

# Print the top 5 most similar institutions (the first one is the Rice University itself)
df.loc[dists.head(6).index]
```

## Question 2

You might be surprised at some of the schools that were found to be "simillar" to Rice just based on admission rate and undergraduate size. Clearly, we need to take some categorical variables into account.

Before we do the full analysis, let's first create a dummy encoding of the two categorical variables:
- the [Carnegie classification](https://en.wikipedia.org/wiki/Carnegie_Classification_of_Institutions_of_Higher_Education)
- the ownership (public, private, etc.)

Calculate the sum over the rows (i.e., the sum of each column). What does this represent?

```python
# do dummy encoding 
pd.get_dummies(df[['CarnegieClassification', 'Ownership']], dtype=int)
```

```python
# Calculate the sum over the dummy variables rows
pd.get_dummies(df[['CarnegieClassification', 'Ownership']], dtype=int).sum(axis=1)     
```

## Question 3

Now let's do the full analysis.

Considering on the admission rate, the number of undergraduates, the Carnegie classification of the type of school, and the ownership (public, private, etc.), what are the 5 schools that are most similar to Rice?

```python
# get the admission rate, the number of undergraduates, dummy variables for Carnegie Classification and Ownership
df_for_full_analysis = df[['AdmissionRate', 'Undergraduates']].join(pd.get_dummies(df[['CarnegieClassification', 'Ownership']], dtype=int), rsuffix='_dummy')
# get standardized values
df_for_full_analysis_standardized = (df_for_full_analysis - df_for_full_analysis.mean(axis='rows')) / df_for_full_analysis.std(axis='rows')
df_for_full_analysis_standardized
```

```python
# find the most similar institutions to Rice University (1432 is the index of Rice University in the original DataFrame)
diffs_for_full_analysis = df_for_full_analysis_standardized - df_for_full_analysis_standardized.loc[1432]
dists_for_full_analysis = np.sqrt((diffs_for_full_analysis ** 2).sum(axis='columns')).sort_values()
dists_for_full_analysis
```

```python
# Print the top 5 most similar institutions (the first one is the Rice University itself)
df.loc[dists_for_full_analysis.head(6).index]
```

## Question 4

The columns whose name begin with "PCIP" contain the proportions of students at each school studying various fields (e.g., Engineering, Psychology). Each field is two-digit code called the [CIP code](https://nces.ed.gov/ipeds/cipcode/browse.aspx?y=55).

What schools are most similar to Rice, if we only consider the proportions of students studying various fields?

```python
# create new dataframe with the CIP code columns
df_cip = df.filter(like='PCIP')
df_cip
```

```python
# get the sum of the CIP code columns
df_cip.sum(axis=1)
```

```python
# find the most similar institutions to Rice University based on CIP codes
# we don't standardize the CIP codes, as they are already on the same scale
diffs_cip = df_cip - df_cip.loc[1432]
dists_cip = np.sqrt((diffs_cip ** 2).sum(axis='columns')).sort_values()
dists_cip
```

```python
# Print the top 5 most similar institutions based on CIP codes
df.loc[dists_cip.head(6).index]
```
